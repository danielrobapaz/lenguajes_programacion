# Analisis comparativo entre las distintas implementaciones de la funcion

$$F_{6,7}(n) =\begin{cases} n &\text{si } 0 \le n \lt 42 \\ F_{6,7}(n-7) + F_{6,7}(n-14) + F_{6,7}(n-21) + F_{6,7}(n-28) + F_{6,7}(n-35) + F_{6,7}(n-42) & \text{si } n \gt 42\end{cases}$$

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_ejecuciones = pd.read_csv('ejecucion_familia_de_funciones.csv')
data_ejecuciones
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=data_ejecuciones, x='n', y='tiempo_recursivo')
plt.title('Tiempo de ejecucion del algoritmo recursivo vs. Tamano de la entrada')
plt.xlabel('Tamano de entrada (n)')
plt.ylabel('Tiempo de ejecucion (segundos)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=data_ejecuciones, x='n', y='tiempo_recursivo_cola')
plt.title('Tiempo de ejecucion del algoritmo recursivo de cola vs. Tamano de la entrada')
plt.xlabel('Tamano de entrada (n)')
plt.ylabel('Tiempo de ejecucion (segundos)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=data_ejecuciones, x='n', y='tiempo_iterativo')
plt.title('Tiempo de ejecucion del algoritmo iterativo vs. tamano de la entrada')
plt.xlabel('Tamano de entrada (n)')
plt.ylabel('Tiempo de ejecucion (segundos)')
plt.show()
```

```python
data_melt = pd.melt(data_ejecuciones, id_vars=['n'], var_name='algoritmo', value_name='tiempo_de_ejecucion')
data_melt['algoritmo'] = data_melt['algoritmo'].apply(lambda a: a[7:].title().replace('_', ' '))
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=data_melt, x='n', y='tiempo_de_ejecucion', hue='algoritmo')
plt.title("Comparacion de los tiempos de ejecucion vs. Tamano de la entrada")
plt.xlabel('Tamano de entrada (n)')
plt.ylabel('Tiempo de ejecucion (segundos)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=data_melt[data_melt['n']<=130], x='n', y='tiempo_de_ejecucion', hue='algoritmo')
plt.title("Comparacion de los tiempos de ejecucion vs. Tamano de la entrada")
plt.xlabel('Tamano de entrada (n)')
plt.ylabel('Tiempo de ejecucion (segundos)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=data_melt[data_melt['algoritmo'].isin(['Recursivo Cola', 'Iterativo'])], x='n', y='tiempo_de_ejecucion', hue='algoritmo')
plt.title("Comparacion de los tiempos de ejecucion vs. Tamano de la entrada")
plt.xlabel('Tamano de entrada (n)')
plt.ylabel('Tiempo de ejecucion (segundos)')
plt.show()
```
